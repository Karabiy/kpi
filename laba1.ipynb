```python
import numpy as np
import unittest

a = [
    [1, 0, -1],
    [0, 4, 4],
    [-1, 4, 14],
]

b = [
    [2],
    [12],
    [19],
]


def makeZeroMatrix(n, m):
    
    
    '''
        Testing whether all parameters are of the correct type
    '''
    
    
    if type(n) != int or type(m) != int:
        raise TypeError('You must give an integer as params')
        
        
    '''
        Testing wheter matrix dimensional is a natural number
    '''
    
    
    if type(n) == int and n < 0 or type(m) == int and m < 0:
        raise ValueError('The number of rows and columns must be a positive number')
        
    '''
        Returning zero matrix with n rows and m columns
    '''
    
    
    return [[0 for i in range(m)] for j in range(n)]


def transposeMatrix(matrix):
    
    
    '''
        Validating matrix - it must be a list or a numpy.ndarray
    '''
    
    
    if type(matrix) not in (list, np.ndarray):
        raise TypeError('You must give a list of lists or a numpy.ndarray as param')
        
    
    '''
        Testing whether the matrix is 2-dimensional
    '''
    
    
    for i in range(len(matrix)):
        if type(matrix[i]) != list and type(matrix[i]) != np.ndarray:
            raise TypeError('The matrix must be 2-dimensional')
    
    
    '''
        Testing whether the matrix has the same count of columns in all rows
    '''
    
    
    columnCount = len(matrix[0])
    for i in range(len(matrix)):
        if len(matrix[i]) != columnCount:
            raise ValueError('The number of columns must be the same in all rows')
            
    
    '''
        Testing whether each of the matrix elements has correct type
    '''
    
    
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if type(matrix[i][j]) not in (int, float, complex):
                raise TypeError('Each of the matrix elements must be an integer, float or complex number')
                
    
    '''
        Returning transposed matrix
    '''
    
    
    transponatedMatrix = makeZeroMatrix(len(matrix[0]), len(matrix))
    for i in range(len(matrix[0])):
        for j in range(len(matrix)):
            transponatedMatrix[i][j] = matrix[j][i]
    return transponatedMatrix


def sqrt(num):
    
    
    '''
        Testing whether parameter has correct type
    '''
    
    
    if type(num) not in (int, float, complex):
        raise TypeError('You must give an integer, float or complex number as param')
    
    
    '''
        Returning square root of given parameter
    '''
    
    
    squareRoot = num**(1/2)
    if squareRoot.imag:
        return round(squareRoot.real, 4) + round(squareRoot.imag, 4) * 1j
    return round(squareRoot.real, 4)


def makeSumForMainDiag(s, i):
    
    
    '''
        Validating matrix - it must be a list or a numpy.ndarray
    '''
    
    
    if type(s) not in (list, np.ndarray):
        raise TypeError('You must give a list of lists or a numpy.ndarray as param')
        
        
    '''
        Testing whether the matrix is 2-dimensional
    '''
    
    
    for j in range(len(s)):
        if type(s[j]) != list and type(s[j]) != np.ndarray:
            raise TypeError('The matrix must be 2-dimensional')
            
    
    '''
        Testing whether the matrix has the same count of columns in all rows
    '''
    
    
    columnCount = len(s[0])
    for _i in range(len(s)):
        if len(s[_i]) != columnCount:
            raise ValueError('The number of columns must be the same in all rows')
            
    
    '''
        Testing whether matrix is squared
    '''
    
    
    if (len(s) != len(s[0])):
        raise ValueError('The matrix must be squared')
    
    
    '''
        Testing whether each of the matrix elements has correct type
    '''
    
    
    for _i in range(len(s)):
        for _j in range(len(s[_i])):
            if type(s[_i][_j]) not in (int, float, complex):
                raise TypeError('Each of the matrix elements must be an integer, float or complex number')
                
    
    '''
        Testing whether i param is natural number
    '''
    
    
    if type(i) != int or i < 0:
        raise TypeError('Param i must be a natural number')
    
    
    '''
        Testing whether i param does not exceed matrix dimension
    '''
    
    
    if i > len(s) + 1:
        raise IndexError('Index exceeds matrix dimension')
        
        
    '''
        Calculating sums for main diag by formula
    '''
    
    
    aggr = 0
    for _i in range(i):
        aggr += s[_i][i] * s[_i][i]
    return aggr


def makeOtherSum(s, i, j):
        
    
    '''
        Validating matrix - it must be a list or a numpy.ndarray
    '''
    
    
    if type(s) not in (list, np.ndarray):
        raise TypeError('You must give a list of lists or a numpy.ndarray as param')
        
        
    '''
        Testing whether the matrix is 2-dimensional
    '''
    
    
    for _j in range(len(s)):
        if type(s[_j]) != list and type(s[_j]) != np.ndarray:
            raise TypeError('The matrix must be 2-dimensional')
            
    
    '''
        Testing whether the matrix has the same count of columns in all rows
    '''
    
    
    columnCount = len(s[0])
    for _i in range(len(s)):
        if len(s[_i]) != columnCount:
            raise ValueError('The number of columns must be the same in all rows')
            
    
    '''
        Testing whether matrix is squared
    '''
    
    
    if (len(s) != len(s[0])):
        raise ValueError('The matrix must be squared')
    
    
    '''
        Testing whether each of the matrix elements has correct type
    '''
    
    
    for _i in range(len(s)):
        for _j in range(len(s[_i])):
            if type(s[_i][_j]) not in (int, float, complex):
                raise TypeError('Each of the matrix elements must be an integer, float or complex number')
                
    
    '''
        Testing whether i and j params is natural number
    '''
    
    
    if type(i) != int or i < 0 or type(j) != int or j < 0:
        raise TypeError('Params i and j must be a natural numbers')
    
    
    '''
        Testing whether i and j params does not exceed matrix dimension
    '''
    
    
    if i > len(s) or j > len(s):
        raise IndexError('Indexes exceeds matrix dimension')

    
    '''
        Calculating sums not for main diag by formula
    '''

    
    aggr = 0
    for _i in range(i+1):
        aggr += s[_i][i] * s[_i][j]
    return aggr


def calculateSMatrix(a):
        
    
    '''
        Validating matrix - it must be a list or a numpy.ndarray
    '''
    
    
    if type(a) not in (list, np.ndarray):
        raise TypeError('You must give a list of lists or a numpy.ndarray as param')
        
        
    '''
        Testing whether the matrix is 2-dimensional
    '''
    
    
    for _j in range(len(a)):
        if type(a[_j]) != list and type(a[_j]) != np.ndarray:
            raise TypeError('The matrix must be 2-dimensional')
            
    
    '''
        Testing whether the matrix has the same count of columns in all rows
    '''
    
    
    columnCount = len(a[0])
    for _i in range(len(a)):
        if len(a[_i]) != columnCount:
            raise ValueError('The number of columns must be the same in all rows')
            
    
    '''
        Testing whether matrix is squared
    '''
    
    
    if (len(a) != len(a[0])):
        raise ValueError('The matrix must be squared')
    
    
    '''
        Testing whether each of the matrix elements has correct type
    '''
    
    
    for _i in range(len(a)):
        for _j in range(len(a[_i])):
            if type(a[_i][_j]) not in (int, float, complex):
                raise TypeError('Each of the matrix elements must be an integer, float or complex number')
    
    
    '''
        Testing whether a is symmetrical
    '''
    
    
    if not np.array_equal(np.array(a), np.array(a).T):
        raise ValueError('The matrix must be symmetrical')
    
    
    '''
        Create empty s matrix
    '''
    
    
    s = makeZeroMatrix(len(a), len(a))
    
    
    '''
        Filling s matrix
    '''
    
    
    for i in range(len(a)):
        for j in range(len(a)):
            if (i == j):
                s[i][i] = sqrt(a[i][i] - makeSumForMainDiag(s, i))
            else:
                s[i][j] = (a[i][j] - makeOtherSum(s, i, j)) / s[i][i] if s[i][i] != 0 else 0
    return s


def getAnswer(s):
        
    
    '''
        Validating matrix - it must be a list or a numpy.ndarray
    '''
    
    
    if type(s) not in (list, np.ndarray):
        raise TypeError('You must give a list of lists or a numpy.ndarray as param')
        
        
    '''
        Testing whether the matrix is 2-dimensional
    '''
    
    
    for _j in range(len(s)):
        if type(s[_j]) != list and type(s[_j]) != np.ndarray:
            raise TypeError('The matrix must be 2-dimensional')
            
    
    '''
        Testing whether the matrix has the same count of columns in all rows
    '''
    
    
    columnCount = len(s[0])
    for _i in range(len(s)):
        if len(s[_i]) != columnCount:
            raise ValueError('The number of columns must be the same in all rows')
            
    
    '''
        Testing whether matrix is squared
    '''
    
    
    if (len(s) != len(s[0])):
        raise ValueError('The matrix must be squared')
    
    
    '''
        Testing whether each of the matrix elements has correct type
    '''
    
    
    for _i in range(len(s)):
        for _j in range(len(s[_i])):
            if type(s[_i][_j]) not in (int, float, complex):
                raise TypeError('Each of the matrix elements must be an integer, float or complex number')
    
    
    '''
        Calculating y matrix
    '''
    
    
    y = np.linalg.solve(np.array(transposeMatrix(s)), b)
    
    
    '''
        Calculating x matrix
    '''
    
    
    x = np.linalg.solve(np.array(s), y)
    return x
    

def solve(a):
    s = calculateSMatrix(a)
    x = getAnswer(s)
    return x
```

```python
class TestServiceMethods(unittest.TestCase):
    
    def testSqrtMethod(self):
        self.assertEqual(sqrt(4), 2.0)
        self.assertEqual(sqrt(0), 0.0)
        self.assertEqual(sqrt(-1), 1j)
        with self.assertRaises(TypeError):
            sqrt('a')
            sqrt(None)

            
    def testMakeZeroMatrixMethod(self):
        self.assertEqual(makeZeroMatrix(0, 0), [])
        self.assertEqual(makeZeroMatrix(1, 1), [[0]])
        self.assertEqual(makeZeroMatrix(2, 2), [[0, 0], [0, 0]])
        self.assertEqual(makeZeroMatrix(3, 1), [
            [0],
            [0],
            [0]
        ])
        with self.assertRaises(TypeError):
            makeZeroMatrix('a', 3)
        with self.assertRaises(ValueError):
            makeZeroMatrix(-5, 5)
    
    
    def testTransposeMatrixMethod(self):
        self.assertEqual(transposeMatrix([
            [1, 2, 2],
            [3, 1, 2],
            [3, 3, 1],
        ]), [
            [1, 3, 3],
            [2, 1, 3],
            [2, 2, 1],
        ])
        self.assertEqual(transposeMatrix([[0]]), [[0]])
        self.assertEqual(transposeMatrix([[1, 2, 3]]), [
            [1],
            [2], 
            [3]
        ])
        with self.assertRaises(TypeError):
            transposeMatrix(0)
            transposeMatrix([0])
            transposeMatrix([
            [1, 2, 2],
            [3, '1', 2],
            [3, 3, 1],
        ])
        with self.assertRaises(ValueError):
            transposeMatrix([
            [1, 2, 2],
            [3, 1],
            [3, 3, 1],
        ])
```

```python
class TestMathMethods(unittest.TestCase):
    
    def testMakeSumForMainDiagMethod(self):
        self.assertEqual(makeSumForMainDiag([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 0), 0)
        self.assertEqual(makeSumForMainDiag([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 1), 0)
        self.assertEqual(makeSumForMainDiag([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 2), 17)
        with self.assertRaises(TypeError):
            makeSumForMainDiag('1', 2)
            makeSumForMainDiag([0], 3)
            makeSumForMainDiag([
                [1, 'b'],
                [2, 3]
            ], 1)
            makeSumForMainDiag([
                [1, 2, 2],
                [3, 1, 2],
                [3, 3, 1],
            ], 'c')
        with self.assertRaises(ValueError):
            makeSumForMainDiag([
                [1, 0, -1],
                [0, 4],
                [-1, 4, 14],
            ], 1)
            makeSumForMainDiag([
                [1, 2, 3],
                [3, 2, 1],
            ], 0)
        with self.assertRaises(IndexError):
            makeSumForMainDiag([
                [1, 2, 2],
                [3, 1, 2],
                [3, 3, 1],
            ], 5)
            
            
    def testMakeOtherSumMethod(self):
        self.assertEqual(makeOtherSum([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 0, 1), 0)
        self.assertEqual(makeOtherSum([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 1, 2), 16)
        self.assertEqual(makeOtherSum([
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ], 2, 1), 72)
        with self.assertRaises(TypeError):
            makeOtherSum('1', 2, 1)
            makeOtherSum([0], 3, 0)
            makeOtherSum([
                [1, 'b'],
                [2, 3]
            ], 1, 1)
            makeOtherSum([
                [1, 2, 2],
                [3, 1, 2],
                [3, 3, 1],
            ], 'c', 'd')
        with self.assertRaises(ValueError):
            makeOtherSum([
                [1, 0, -1],
                [0, 4],
                [-1, 4, 14],
            ], 1, 2)
            makeOtherSum([
                [1, 2, 3],
                [3, 2, 1],
            ], 0, 1)
        with self.assertRaises(IndexError):
            makeOtherSum([
                [1, 2, 2],
                [3, 1, 2],
                [3, 3, 1],
            ], 5, 9)
            
            
    def testCalculateSMatrixMethod(self):
        matrix = [
            [1, 0, -1],
            [0, 4, 4],
            [-1, 4, 14],
        ]
        answer = [[1.0, 0.0, -1.0], [0, 2.0, 2.0], [0, 0, 3.0]]

        self.assertEqual(calculateSMatrix(matrix), answer)
        with self.assertRaises(TypeError):
            calculateSMatrix('1')
            calculateSMatrix([0])
            calculateSMatrix([
                [1, 'b'],
                [2, 3]
            ])
        with self.assertRaises(ValueError):
            calculateSMatrix([
                [1, 0, -1],
                [0, 4],
                [-1, 4, 14],
            ])
            calculateSMatrix([
                [1, 0],
                [0, 4],
                [1, 4],
            ])     
            calculateSMatrix([
                [1, 0, -1],
                [0, 4, 5],
                [-1, 4, 14],
            ])
    def testGetAnswerMethod(self):
        matrix = [
            [1.0, 0.0, -1.0],
            [0, 2.0, 2.0],
            [0, 0, 3.0],
        ]
        answer = np.array([
            [3.],
            [2.],
            [1.]]
        )
        isAnswerCorrect = np.array_equal(getAnswer(matrix), answer)
        self.assertEqual(isAnswerCorrect, True)
        with self.assertRaises(TypeError):
            getAnswer('1')
            getAnswer([0])
            getAnswer([
                [1, 'b'],
                [2, 3]
            ])
        with self.assertRaises(ValueError):
            getAnswer([
                [1, 0, -1],
                [0, 4],
                [-1, 4, 14],
            ])
            getAnswer([
                [1, 0],
                [0, 4],
                [1, 4],
            ]) 
```
